```python
days=[1, 2, 3, 4, 5]
sleepings = [7, 6, 8, 7, 5]
eatings = [2, 1, 3, 2, 1]
working = [9, 10, 12, 11, 9]
playing = [6, 5, 5, 3, 4]
pl.stackplot(days, sleepings, eatings, working, colors=['blue', 'orange', 'red'])
pl.show()

pl.pie( playing, labels=["days", "sleepings", "eatings", "working", "playing"], colors=['r', 'g', 'b', 'y', 'black'], startangle=100)
pl.show()
```

```python
salary_data  = pd.read_csv('./salaryData.csv')
salary_data
```

```python
salary_data[ salary_data['Salary'] > salary_data['Salary'].mean()]
```

```python
salary_data.groupby(['YearsExperience']).sum()
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
x_train, x_test, y_train, y_test = train_test_split(salary_data['YearsExperience'], salary_data['Salary'], test_size=0.3, random_state=10)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
pl.scatter(x_train, y_train, color='r' )
```

```python
from sklearn.linear_model import LinearRegression

clf = LinearRegression()
clf.fit(x_train.to_frame(), y_train.to_frame())
```

```python
print(x_test)
print(y_test)
clf.predict(x_test.to_frame())
```

```python
x_test = x_test.to_frame()
y_test = y_test.to_frame()
clf.score(x_test, y_test)
```

```python
clf.coef_
```

```python
clf.intercept_
```

```python
# y = 9303.95 * x + 27206.42
```

```python
from sklearn.metrics import mean_squared_error
acc = mean_squared_error(x_test, y_test)
```

```python
acc
```

```python
y_test_predict = clf.predict(x_test)
pl.scatter(x_test, y_test)
pl.plot(x_test, y_test_predict)
pl.show()
```
