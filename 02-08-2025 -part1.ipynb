```python
import pandas as pd
import sklearn
```

```python
data = pd.read_csv('./home.csv')
data
```

```python
data['bedroom'] = data['bedroom'].fillna(data['bedroom'].mean())
data
```

```python
data['bathroom'] = data['bathroom'].fillna(data['bathroom'].mean())
data['balcony'] = data['balcony'].fillna(data['balcony'].mean())
data
```

```python
y = data['price']
x = data.drop('price', axis=1)
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=5)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn import linear_model

ml = linear_model.LinearRegression()
ml.fit(x_train, y_train)
```

```python
y_predict = ml.predict(x_test)
y_predict
```

```python
from word2number import w2n

data['Roll_no'] = data['Roll_no'].apply(w2n.word_to_num)
data
```

```python
data['Written_marks'] = data['Written_marks'].fillna('0')
data
```

```python
y = data['total_marks']
x = data.drop('total_marks', axis=1)
y
```

```python
ml2 = sklearn.linear_model.LinearRegression()
ml2.fit(x, y)
```

```python
ml2.predict([[12, 5, 5]])
```

```python
# saving the model
# pickle
# sklearn.joblib
```

```python
# pickle - to convert python object into the serializer (bytes code)
# complex models - joblib - d
```

```python
import pickle

with open('marks_model', 'wb') as fp:
    pickle.dump(ml2, fp)
```

```python
with open('marks_model', 'rb') as fp:
    marks_model = pickle.loads(fp.read())
print(marks_model)
```

```python
marks_model.predict([[55, 6, 7]])
```
