```python
import pandas as pd
import sklearn
```

```python
data = pd.read_csv('./home.csv')
data
```

```python
data['bedroom'] = data['bedroom'].fillna(data['bedroom'].mean())
data
```

```python
data['bathroom'] = data['bathroom'].fillna(data['bathroom'].mean())
data['balcony'] = data['balcony'].fillna(data['balcony'].mean())
data
```

```python
y = data['price']
x = data.drop('price', axis=1)
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=5)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn import linear_model

ml = linear_model.LinearRegression()
ml.fit(x_train, y_train)
```

```python
y_predict = ml.predict(x_test)
y_predict
```
