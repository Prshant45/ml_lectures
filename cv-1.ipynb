```python
import cv2

print(cv2.__version__)
```

```python
image = cv2.imread('building.jpg')
```

```python
from matplotlib import pyplot as plt

def print_image(image, title="hello", size=10):
    h, l = image.shape[0], image.shape[1]
    aspect_ratio =  h /l
    plt.figure(figsize=(size*aspect_ratio, size))
    # plt.imshow(image)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
   # plt.imshow(cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
   # plt.imshow(cv2.cvtColor(image, cv2.COLOR_RGB2GRAY))
    plt.title(title)
    plt.show()

print_image(image)
```

```python
image = np.zeros((512, 512, 3), dtype="uint8")
```

```python
cv2.line(image, (0,0), (511, 511), (255, 127, 0), 5)
plt.imshow(image)
```

```python
image = np.zeros((512, 512, 3), dtype="uint8")
cv2.rectangle(image, (100, 100), (300, 250), (255, 127, 0), 5)
plt.imshow(image)
```

```python
image = np.zeros((512, 512, 3), dtype="uint8")
cv2.circle(image, (250, 250), 100, (255, 127, 0), -1)
plt.imshow(image)
```

```python
image = np.zeros((512, 512, 3), dtype="uint8")
pts = np.array([[10, 50], [400, 50], [90, 100], [50, 500]], np.int32)
cv2.polylines(image, [pts], True, (255, 127, 0), 3)
plt.imshow(image)
```

```python
image = np.zeros((1200, 1200, 3), dtype="uint8")
print_string = "Hello World!"
cv2.putText(image, print_string, (100, 200), cv2.FONT_HERSHEY_SIMPLEX, 2, (40, 200, 0), 5)
plt.imshow(image)
```
