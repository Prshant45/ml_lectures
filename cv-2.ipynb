```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
    
```

```python
def show_image(image, title="image", size=10): 
    h, l = image.shape[0], image.shape[1]
    aspect_ratio =  h /l
    plt.figure(figsize=(size*aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
image = cv2.imread('building.jpg')
show_image(image, "building", 10)
```

```python
def transpose(image):
    heigth, width = image.shape[:2]
    T = np.float32([[1, 0, heigth/4], [0, 1, width/4]])
    image_translation = cv2.warpAffine(image, T, (width, heigth))
    show_image(image_translation, title="translation")
transpose(image)
```

```python
def rotation(image):
    heigth, width = image.shape[:2]
    print(heigth, width)
    # (width/2, heigth/2) - identify the centre
    rotate = cv2.getRotationMatrix2D((width/2, heigth/2), 180, 1)
    image_rotation = cv2.warpAffine(image, rotate, (width, heigth))
    show_image(image_rotation, title="rotation")
rotation(image)
```
